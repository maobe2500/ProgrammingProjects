```python
import cmath
%matplotlib inline
from matplotlib import pyplot as plt

a = -0.74
b = 0.1


def escape(a,b):

    c = complex(a,b)
    z_list = []

    for i in range(10):
        if (i == 0):
            z = 0 + c
        else:
            z = z ** 2 + c
        z_list.append(z)

    return z_list



def distance_to_origin(a,b):

    c = complex(a,b)
    dist_list = []

    for i in range(100):
        if (i == 0):
            z = 0 + c
        else:
            z = z ** 2 + c
        
        dist = abs(z)
        dist_list.append(dist)

    return dist_list

print(distance_to_origin(a,b))



x = [i for i in range(100)]
y = distance_to_origin(a,b)

fig = plt.figure()
ax = fig.add_axes([0,0,1.5,1.5])
ax.set_title("")
ax.set_xlabel("Antal iterationer")
ax.set_ylabel("Avstånd från origo")
plt.rc('axes', labelsize=15)
plt.plot(x,y)
plt.axis([0, 100, -1, 5])
plt.show()
```

```python
import numpy as np
import matplotlib as plt
%matplotlib inline

def logistic(r, x):
    return r * x * (1 - x)

x = np.linspace(0, 1)
fig, ax = plt.pyplot.subplots(1, 1)
ax.plot(x, logistic(2, x), "k")
```
